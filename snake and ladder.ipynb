```python
import random     


player1=random.randint(1,6)
player2=random.randint(1,6)
pos1 = 0
pos2 = 0
ladder = {4: 14, 8: 30, 21: 42, 28: 76, 50: 67, 71: 92, 88: 99}
snake = {32: 10, 36: 6, 28: 26, 62: 18, 88: 24, 95: 56, 97: 78}

def game(pos):
    dice = random.randint(1, 6)
    input("dice: {}".format(dice))
    pos = pos + dice
    
    if pos in snake:
        print("bitten by snake")
        pos = snake[pos]
        print("position: {}".format(pos))
    elif pos in ladder:
        print("climb by ladder")
        pos = ladder[pos]
        print("position: {}".format(pos))
    elif pos >= 100:
        print(" winner of the game")
        
    else:
        print("position: {}".format(pos))
        print("\n")        
    return pos

while True:
    input("player1 throw a dice: ")
    pos1 = game(pos1)
    if pos1 >= 100:
        print("game over")
        print("player 1 winner")
        break
        
    input("player2 throw a dice: ")
    pos2 = game(pos2)
    if pos2 >= 100:
        print("game over")
        print("player 2 winner")
        break

        
```
